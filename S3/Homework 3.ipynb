## Задание 1
### Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150.
### Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
from math import sqrt

arr = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]

# вычесляем среднее арифметическое
def mean (a : list) -> float:
    return sum(a)/len(a)

# вычесляем смещенную дисперсию
def var (a : list) -> float:
    sum = 0
    mean_num = mean(a)
    for i in a:
        sum += (i - mean_num)**2
    return sum / len(a)

# вычесляем несмещенную дисперсию
def var1 (a : list) -> float:
    sum = 0
    mean_num = mean(a)
    for i in a:
        sum += (i - mean_num)**2
    return sum / (len(a) - 1)



print(f'Среднее арифмитическое равно: {mean(arr)}')
print(f'Смещенная дисперсия равна: {round(var(arr), 3)}')
print(f'Смещенное среднее квадратичное отклонение: {round(sqrt(var(arr)), 3)}')
print(f'Несмещенная дисперсия равна: {round(var1(arr), 3)}')
print(f'Несмещенное среднее квадратичное отклонение: {round(sqrt(var1(arr)), 3)}')
```
