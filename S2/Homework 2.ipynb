# Практическое задание к 2 семинару (Дискретные распределения вероятностей)
## Задание 1
***Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз.***

***Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.***

$p = 0,8$ , $n = 100$ , $m = 85$

Используем формулу Бернули $P_{100}^{85}=C_{100}^{85}*0.8^{85}*(1-0.8)^{100-85}$

**Ответ: $P_{100}^{85}=0.04806$**

## Задание 2
***Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек.***

n = 5000

e = 2.74

$\Lambda = n * p = 5000 * 0.0004 = 2$

***а) Какова вероятность, что ни одна из них не перегорит в первый день?***

m = 0

Используем распределение Пуассона  $P_{5000}^{0}=\frac{2^0}{0!}*e^-2 = 0.1353 $

***б) Какова вероятность, что перегорят ровно две?***

m = 2

Используем распределение Пуассона  $P_{5000}^{2}=\frac{2^2}{2!}*e^-2 = 0.2707 $

**Ответ: $P_{5000}^{0} = 0.1353 ,   P_{5000}^{2} = 0.2707$**

### Задание 3
***Монету подбросили 144 раза.***

***Какова вероятность, что орел выпадет ровно 70 раз?***

n = 144

p = 0.5

m = 70

Используем формулу Бернули $P_{144}^{70}=C_{144}^{70}*0.5^{70}*0.5^{144-70} = 0.0628$

**Ответ: $P_{144}^{70} = 0.0628$**


### Задание 4
***В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых.***

***Из каждого ящика вытаскивают случайным образом по два мяча.***



***а) Какова вероятность того, что все мячи белые?***

$\frac{7}{10} * \frac{6}{9} * \frac{9}{11} * \frac{8}{9} = 0,3055$

***б)Какова вероятность того, что ровно два мяча белые?***

Находим вероятность что из первого ящика достали 2 белых мяча а из второго неодного

$\frac{7}{10} * \frac{6}{9} * \frac{2}{11} * \frac{1}{10} = 0,008388$

Находим вероятность что из первого ящика достали неодного белого мяча а из второго 2 

$\frac{3}{10} * \frac{2}{9} * \frac{9}{11} * \frac{8}{10} = 0,043623$

Находим вероятность что из первого ящика достали 1 белый мяч и из второго 1

$\frac{7}{10} * \frac{3}{9} * \frac{9}{11} * \frac{2}{10} = 0,1525$

Сложим получившиеся вероятности так как наступит **ИЛИ** одно событие **ИЛИ** другое **ИЛИ** третье.

0,008388 + 0,043623 + 0,1525 = 0,2045


***в) Какова вероятность того, что хотя бы один мяч белый?***

Найдём вероятность обратную (что неодного белого шара не выпало)

для первого ящика $\frac{3}{10} * \frac{2}{9} = \frac{1}{15}$

для второго ящика $\frac{2}{11} * \frac{1}{10} = \frac{1}{55}$

$\frac{1}{15} * \frac{1}{55} = 0.0012$

Соответственно вероятность выпадения хотябы 1 белого шара равна 1 - 0.0012 = 0,9988

**Ответ: а) 0,3055 б) 0,2045 в) 0,9988**
